```python
for ind in set(df['Industry']):
    df[df['Industry'] == ind].plot(x = 'FromIPO', y = 'Return', kind = 'scatter', title = ind)
```

```python
mC = list()
aV = list()
del_list = list()
for tick in df['Symbol']:
    try: 
        inf = yf.Ticker(tick)
        mC.append(inf.info['marketCap'])
        aV.append(inf.info['averageVolume'])
    except: del_list.append(tick)
```

```python
for tick in del_list:
    df.drop(((df.index[df['Symbol'] == tick].tolist()[0])), inplace = True)
```

```python
df['Market Cap'] = mC
df['Avg Volume'] = aV
```

```python
df[df['Industry']== 'Technology'].corr()
df.dropna(inplace = True)
```

```python
linreg = LinearRegression()
def build_model(seed):
    import numpy as np
    from sklearn.linear_model import LinearRegression
    from sklearn.model_selection import train_test_split

    X = df[['1st Day Jump', 'Market Cap', 'Avg Volume']]
    y = df[['Return']]

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=seed)


    linreg.fit(X_train, y_train)

    # prediction
    y_pred = linreg.predict(X_test)
    y_pred = [i[0] for i in y_pred]
    
    #build out df1
    df1 = pd.DataFrame()
    df1['yPred'] = y_pred
    df1['yTest'] = list(y_test['Return'])
    df1.index = y_test.index
    df1['Residual'] = df1['yTest'] - df1['yPred'] 

    return df1
```

```python
days_ = []
def build_investments(df1):
    inv = dict()
    for idx, ele in df1.sort_values(by = 'yPred', ascending = False).head(10).iterrows():
        buy = df.loc[idx]['1st Day Close']
        sell = df.loc[idx]['Current Price']
        days_.append(df.loc[idx]['FromIPO'])
        inv[df.loc[idx]['Symbol']] = sell/buy
    
    return inv
```

```python
for i in range(1):
    df1 = build_model(i)
    inv = build_investments(df1)
    print(np.mean(list(inv.values())))
```
